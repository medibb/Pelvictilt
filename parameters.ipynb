```python
import numpy as np
from scipy.ndimage import gaussian_filter1d

dataset = 'sst220202_a'

def inflection_detection(data):
    dd = gaussian_filter1d(data, 1)
    ext_dd = np.gradient(dd)
    infls = np.where(np.diff(np.sign(ext_dd)))[0]
    return infls


## 데이터 경로 지정
data = np.load('./npy/' + dataset + '_data.npy')
    
## 라벨 경로 지정
label = np.load('./npy/' + dataset + '_label.npy')
    

     
for i in range(len(label)):
# for i in range(40, len(label)):  # 오류 번호부터 다시
    inf = inflection_detection(data[i, 1, 2])  #sacrum sensor
    if inf[0] <= label[i] :       #start가 존재해야함
        start = np.where(inf <= label[i])[0][-1]
        finish = np.where(inf >= label[i])[0][-0]
        PPA = data[i, 1, 2, inf[finish]] - data[i, 1, 2, inf[start]]
    else :
        start = np.where(inf)[0][0]
        finish = np.where(inf)[0][1]
        PPA = data[i, 1, 2, inf[finish]] - data[i, 1, 2, inf[start]]
        print('**%d번째 start point 설정 오류**' % i)
        
    KA = data[i, 2, 2] + data[i, 3, 2]
    LL = 180 - data[i, 0, 2] + data[i, 1, 2]
    
    over180 = np.where((LL - 180) > 0)[0] #idx 만
    under180 = np.where((LL - 180) < 0)[0]
    
    
    if len(under180) == 0 or len(over180) == 0 :    #over180 or under 180 값이 없는 경우
        if PPA <= 0 :
            print( '(%d번째*+)' % i, inf[start], inf[finish], 'NA', 'NA', round(label[i]), "%0.2f" % PPA)
            print('inflection points of pelvis:', inf)
        else:
            print( '(%d번째+)' % i, inf[start], inf[finish], 'NA', 'NA', round(label[i]), "%0.2f" % PPA)
    else:                                          #over180 값이 하나라도 있는 경우
        if under180[0] <= inf[start] :  #첫번째 180이하 값이 SPPT보다 작으면 정상 LL
            over180 = over180[over180 > inf[start]]
            if PPA <= 0 :
                print( '(%d번째*)' % i, inf[start], inf[finish], over180[0], over180[-1], round(label[i]), "%0.2f" % PPA)
                print('inflection points of pelvis:', inf)
            else:
                print( '(%d번째)' % i, inf[start], inf[finish], over180[0], over180[-1], round(label[i]), "%0.2f" % PPA)
        else:                         #비정상 LL
            if PPA <= 0 :
                print( '(%d번째+*)' % i, inf[start], inf[finish], 'NA', over180[-1], round(label[i]), "%0.2f" % PPA)
                print('inflection points of pelvis:', inf)
                print('Lordosis under 180:',under180)
            else:
                print( '(%d번째+)' % i, inf[start], inf[finish], 'NA', over180[-1], round(label[i]), "%0.2f" % PPA)
                print('Lordosis under 180:',under180)

print('initial point of SPPT/', 'end point of SPPT/', 'initial point of SLL/', 'end point of SLL/', 'but wink', 'PPA')
```
